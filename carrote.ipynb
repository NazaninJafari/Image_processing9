```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('imgs/havich.jpg')
image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
plt.imshow(image)

print(image.shape)
```

```python
R,G,B = cv2.split(image)

R_new = G
G_new = G
B_new = B

result = cv2.merge((R_new, G_new, B_new))
```

```python
plt.imshow(result)
result = cv2.cvtColor(result, cv2.COLOR_BGR2RGB)
cv2.imwrite('output_imgs/result.jpg', result)
```
