```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('imgs/mokaab.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
w,h,_ = image.shape

for i in range(w):
    for j in range(h):
        R,G,B = image[i,j]
        if R<40 and G>200 and B<40:
            R=255
            G=0
            B=255
        elif R<40 and G<40 and B>200:
            R=255
            G=255
            B=0
        elif R>200 and G<40 and B<40:
            R=0
            G=255
            B=255
        image[i,j]= R,G,B     
```

```python
plt.imshow(image) 
cv2.imwrite('output_imgs/rubik_result.jpg',image)
```
